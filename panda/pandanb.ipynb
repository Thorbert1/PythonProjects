Assuming the time taken for my jump was 0.7 seconds, and gravity on earth is 9.81 $ m / s^2 $, the height I jumped is:

```python
GRAVITY = 9.81
deltaTime = 0.18/0.30
halfOfDelta = deltaTime/2

height = (1/2) * (GRAVITY * halfOfDelta * halfOfDelta)
print(f"{height:.3f}m, which is roughly equal to {(height * 39.3701):.2f}in")
```

Inversely, if I wanted to verify that someone has a 50in vertical leap, then they'd have to be able to hang in the air for:

```python
import math

GRAVITY = 9.81
heightInInches = 50
heightInMeters = heightInInches/39.3701

print(f"{math.sqrt(2 * heightInMeters / GRAVITY) * 2:.3f}s")
```
